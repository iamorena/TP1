```python
# 1) detectar la columna 'region' sin cambiar nombres
def find_region_col(df): #Recorre el nombre de las columnas del DataFrame
    for c in df.columns:
        if str(c).strip().lower() == "region":
            return c
    raise KeyError("No encuentro la columna 'region'.")

#A cada columna lo convierte a string, le saca espacios con el strip() y lo
# pasa a minuscula
# esta función busca cómo se llama exactamente la columna región en la tabla
# recorre todas las columnas utilizando el dor y a cada nombre lo pasa a minusula y
# le saca los espacios; si queda "region" devuelve ese nombre original c.

colR_2005 = find_region_col(base2005)
colR_2025 = find_region_col(base2025)

# esto guarda el nombre real de la columna region en cada base

# 2) máscara para GBA (01)
def mask_gba(df, col):
    s = df[col]
    return (
        pd.to_numeric(s, errors="coerce").eq(1) |
        s.astype(str).str.strip().str.zfill(2).eq("01") |
        s.astype(str).str.strip().str.lower().isin(["gran buenos aires","g.b.a","gba"])
    )
# Esta función devuelve True las filas que son de GBA y False e las que no
# si la región viene con un 1, 01 o con una palabra  la marca como GBA
# usamos el astype("string") para que no se rompa si hay numeros o NaN mezclados.


# 3) filtrar SIN tocar ninguna otra columna
base2005_gba = base2005.loc[mask_gba(base2005, colR_2005)].copy()
base2025_gba = base2025.loc[mask_gba(base2025, colR_2025)].copy()

# El .loc[mask] se queda con las filas True de arriba y hace una copia

# 4) verlas “como el original”, solo filtradas
base2005_gba.head()
base2025_gba.head()

base2005_gba.to_csv("base2005_gba.csv")
base2025_gba.to_csv("base2025_gba.csv")
```

2c. Corregir variables si notan valores sin sentido (como ingresos negativos) de acuerdo a la documentación de la EPH (puede ser una codificación de no respuesta de los individuos) y eliminen estos valores extraños de sus 15 variables de interés. Comenten brevemente en el reporte dicho proceso de limpieza.

```python
import pandas as pd
import numpy as np
import unicodedata

# === Partimos de la base original (SIN renombres)
df = base_gba15.copy()

# ---------- NORMALIZACIÓN PREVIA DE ch04 (SEXO) ----------
def _strip_accents(s: str) -> str:
    return ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))

def normalizar_sexo(col):
    """
    Devuelve serie numérica: 1=Varón / 2=Mujer, a partir de
    valores 1/2 (numéricos o strings) o descripciones textuales.
    """
    s = col.copy()

    # a) Pasar todo a string, trim, minúscula y sin acentos
    s = s.astype(str).str.strip()
    s = s.replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})
    s_norm = s.dropna().map(lambda x: _strip_accents(str(x)).lower())

    out = pd.Series(pd.NA, index=col.index, dtype="Int64")

    # b) Casos ya codificados como 1/2 (incluyendo strings "1","2")
    mask_1 = s.isin([1, "1"])
    mask_2 = s.isin([2, "2"])
    out.loc[mask_1] = 1
    out.loc[mask_2] = 2

    # c) Textuales
    if not s_norm.empty:
        # varón / hombre / masculino
        mask_varon = s_norm.str.contains(r"\b(varon|hombre|mascul)\b", regex=True)
        # mujer / femenino
        mask_mujer = s_norm.str.contains(r"\b(mujer|femen)\b", regex=True)
        out.loc[mask_varon.index[mask_varon]] = 1
        out.loc[mask_mujer.index[mask_mujer]] = 2

    return out

# Si existe ch04, la normalizamos ANTES de cualquier otra limpieza
if "ch04" in df.columns:
    df["ch04"] = normalizar_sexo(df["ch04"])

# --- Reglas conservadoras (EPH) ---
# 1) INGRESOS: negativos -> NaN
ingresos_vars = [c for c in ["ipcf", "itf"] if c in df.columns]

# 2) EDAD (ch06): fuera de [0, 110] -> NaN
edad_var = "ch06" if "ch06" in df.columns else None

# 3) BINARIAS 1/2: fuera de {1,2} -> NaN
binarias_12 = [c for c in ["ch09", "pp02i", "ch10", "pp06a", "ch08"] if c in df.columns]

# 4) SEXO (ch04): fuera de {1,2} -> NaN  (ya normalizada arriba, acá sólo validamos)
sexo_var = "ch04" if "ch04" in df.columns else None

# 5) OTRAS CATEGÓRICAS: negativos -> NaN (regla mínima segura)
otras_categoricas = [c for c in ["nivel_ed","estado","cat_inac","pp04a","pp04g"] if c in df.columns]

# --- Copia para limpiar y log de cambios
subset_clean = df.copy()
changes = {}

# 1) Ingresos
for v in ingresos_vars:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v] < 0
    changes[f"{v}_negativos"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# 2) Edad
if edad_var:
    subset_clean[edad_var] = pd.to_numeric(subset_clean[edad_var], errors="coerce")
    mask = (subset_clean[edad_var] < 0) | (subset_clean[edad_var] > 110)
    changes[f"{edad_var}_fuera_rango"] = int(mask.sum())
    subset_clean.loc[mask, edad_var] = pd.NA

# 3) Binarias 1/2
for v in binarias_12:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v].notna() & ~subset_clean[v].isin([1,2])
    changes[f"{v}_fuera_1_2"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# 4) Sexo 1/2 (validación)
if sexo_var:
    # ya está en Int64 por la normalización; sólo validamos
    mask = subset_clean[sexo_var].notna() & ~subset_clean[sexo_var].isin([1,2])
    changes[f"{sexo_var}_fuera_1_2"] = int(mask.sum())
    subset_clean.loc[mask, sexo_var] = pd.NA

# 5) Otras categóricas (negativos)
for v in otras_categoricas:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v] < 0
    changes[f"{v}_negativos"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# --- Resultado final
base_gba15_clean = subset_clean #armo una nueva base de datos con las variables limpias

print("Recodificaciones efectuadas (celdas afectadas):")
for k, v in changes.items():
    if v > 0:
        print(f"  - {k}: {v}")
print(f"\nBase limpia -> base_gba15_clean  |  shape: {base_gba15_clean.shape}")

# (chequeo rápido opcional para que veas que NO quedan NaN masivos en 2005)
if "ano4" in base_gba15_clean.columns and "ch04" in base_gba15_clean.columns:
    print("\nDistribución ch04 por año (post-limpieza):")
    print(base_gba15_clean.groupby("ano4")["ch04"].value_counts(dropna=False))

# Guardar
base_gba15_clean.to_csv("base_gba15_clean.csv", index=False)
```

```python
# === Matriz de correlación con 7 variables (base_gba15 en minúsculas) ===
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt

assert 'base_gba15' in globals(), "No encuentro 'base_gba15' en memoria."

# Columnas tal como están en base_gba15
vars_corr = ["ch04","ch06","ch07","ch08","nivel_ed","estado","ipcf"]
vars_all  = vars_corr + ["ano4"]

# Selección defensiva (por si faltara alguna)
cols_exist = [c for c in vars_all if c in base_gba15.columns]
df = base_gba15[cols_exist].copy()

# Tipos: numéricas vs categóricas (dejando NaN como NaN)
if "ch06" in df.columns:
    df["ch06"] = pd.to_numeric(df["ch06"], errors="coerce")
if "ipcf" in df.columns:
    df["ipcf"] = pd.to_numeric(df["ipcf"], errors="coerce")

for c in ["ch04","ch07","ch08","nivel_ed","estado"]:
    if c in df.columns:
        codes = df[c].astype("category").cat.codes.replace(-1, pd.NA)
        df[c] = pd.to_numeric(codes, errors="coerce")

# Nombres lindos SOLO para el gráfico (no modifico la base original)
nombres = {
    "ch04": "Sexo",
    "ch06": "Edad",
    "ch07": "Estado civil",
    "ch08": "Cobertura médica",
    "nivel_ed": "Nivel educativo",
    "estado": "Cond. de actividad",
    "ipcf": "Ing. per cápita familiar",
    "ano4": "Año"
}
df_plot = df.rename(columns=nombres)

def plot_corr(df_labeled, year, decimals=2):
    sub = df_labeled[df_labeled["Año"] == year].copy()
    cols7 = [nombres[v] for v in vars_corr if nombres[v] in sub.columns]
    X = sub[cols7]

    # descartar columnas sin varianza (≤1 valor no nulo distinto)
    keep = [c for c in X.columns if X[c].dropna().nunique() > 1]
    dropped = [c for c in X.columns if c not in keep]
    if dropped:
        print(f"Año {year}: columnas sin varianza descartadas -> {dropped}")
    X = X[keep]

    corr = X.corr(numeric_only=True).round(decimals)

    # Heatmap
    n = corr.shape[0]
    side = max(8, 1.2*n)
    sns.set_context("notebook"); sns.set(font_scale=0.9)
    plt.figure(figsize=(side, side))
    sns.heatmap(
        corr, cmap="coolwarm", center=0, vmin=-1, vmax=1,
        square=True, linewidths=.3,
        annot=True, fmt=f".{decimals}f", annot_kws={"size": 9},
        cbar_kws={'label':'ρ de Pearson'}
    )
    plt.title(f"Matriz de correlación - Año {year}", fontsize=16)
    plt.xticks(rotation=45, ha="right"); plt.yticks(rotation=0)
    plt.tight_layout(); plt.show()

    # devolver matriz
    return corr

# Ejecutar para 2005 y 2025
corr_2005 = plot_corr(df_plot, 2005, decimals=2)
corr_2025 = plot_corr(df_plot, 2025, decimals=2)

# === Imprimir tablas de correlación ===
print("\n=== Tabla de correlación - Año 2005 ===")
display(corr_2005)

print("\n=== Tabla de correlación - Año 2025 ===")
display(corr_2025)
```

```python
import pandas as pd
import numpy as np

# 1) Cargamos respondieron.csv
df = pd.read_csv("respondieron.csv", low_memory=False)
df.columns = df.columns.str.strip().str.lower()

# Detectar columnas sexo/edad en respondieron
sexo_col = "ch04" if "ch04" in df.columns else "sexo"
edad_col = "ch06" if "ch06" in df.columns else "edad"

# 2) Cargar tabla_adulto_equiv.xlsx
tabla = pd.read_excel("/content/ADULTO_EQUIVALENTE.xlsx", sheet_name=0, header=0)
# header=3 porque los títulos reales empiezan en la fila 4 de tu captura
tabla.columns = tabla.columns.str.strip().str.lower()

# Ahora tabla tiene: 'edad' | 'mujeres' | 'varones'

# 3) Pasar de ancho a largo
tabla_largo = tabla.melt(id_vars=["edad"], value_vars=["mujeres","varones"],
                         var_name="sexo_txt", value_name="adulto_equiv")

# Mapear sexo a números EPH (1=varón, 2=mujer)
mapa_sexo = {"varones":1, "mujeres":2}
tabla_largo["sexo"] = tabla_largo["sexo_txt"].map(mapa_sexo)

# Normalizar edad → tomar solo números (primer número de la celda)
def extraer_edad(e):
    if pd.isna(e): return np.nan
    s = str(e).lower()
    nums = [int(x) for x in s.split() if x.isdigit()]
    return nums[0] if nums else np.nan

tabla_largo["edad_num"] = tabla_largo["edad"].map(extraer_edad).astype("Int64")

tabla_equiv = tabla_largo[["sexo","edad_num","adulto_equiv"]].dropna()

# 4) Preparar df respondieron
df[sexo_col] = pd.to_numeric(df[sexo_col], errors="coerce")
df[edad_col] = pd.to_numeric(df[edad_col], errors="coerce").astype("Int64")

# 5) Merge por sexo + edad
df = df.merge(
    tabla_equiv,
    how="left",
    left_on=[sexo_col, edad_col],
    right_on=["sexo","edad_num"]
).drop(columns=["sexo","edad_num"], errors="ignore")

# 6) Calcular suma por hogar
hogar_keys = [c for c in ["codusu","nro_hogar","hogar"] if c in df.columns]
df["ad_equiv_hogar3"] = df.groupby(hogar_keys)["adulto_equiv"].transform("sum")

# 7) Guardar
df.to_csv("respondieron_enriquecido.csv", index=False)
print("Archivo guardado: respondieron_enriquecido.csv")

# Vista rápida
cols_show = hogar_keys + [sexo_col, edad_col, "adulto_equiv", "ad_equiv_hogar3"]
display(df.loc[df["adulto_equiv"].notna(), cols_show])
```

7.Sabiendo que la Canasta Básica Total para un adulto equivalente en
el primer trimestre de 2025 es aproximadamente $365.177, agreguen

ingreso_necesario que sea el producto de este valor por
ad_equiv_hogar. Para el primer trimestre de 2005 la Canasta
Básica Total para un adulto equivalente era aproximadamente
$205,07. Note que este es el valor mínimo que necesita ese hogar para
no ser pobre.

```python
import pandas as pd
import numpy as np

# === 1) Cargar base enriquecida ===
df = pd.read_csv("respondieron_enriquecido.csv", low_memory=False)

# Me aseguro que ano4 esté numérico
df["ano4"] = pd.to_numeric(df["ano4"], errors="coerce")

# === 2) Definir valores de la canasta básica por adulto equivalente ===
canasta = {2005: 205.07, 2025: 365177}

# === 3) Calcular ingreso necesario ===
df["ingreso_necesario"] = df.apply(
    lambda r: canasta.get(int(r["ano4"]), np.nan) * r["ad_equiv_hogar3"]
    if pd.notna(r["ano4"]) and pd.notna(r["ad_equiv_hogar3"]) else np.nan,
    axis=1
)

# === 4) Corregir ceros falsos (cambiar por NaN) ===
df.loc[df["ad_equiv_hogar3"] == 0, ["ad_equiv_hogar3", "ingreso_necesario"]] = np.nan

# === 5) Guardar base corregida ===
df.to_csv("respondieron_p7.csv", index=False)
print("✔ Guardado respondieron_p7.csv")

# === 6) Info de chequeo ===
print("\nHogares con NaN en ad_equiv_hogar3:", df["ad_equiv_hogar3"].isna().sum())
print("\nDistribución ingreso_necesario por año:")
print(df.groupby("ano4")["ingreso_necesario"].describe())

# === 7) Mostrar tabla estilo la que tenías con 0.000 ===
print("\n=== Vista previa de la tabla corregida ===")
print(df[["ano4","adulto_equiv", "ad_equiv_hogar3", "ingreso_necesario"]].head(20))

# me tira NaN los que no se pudo calcular el aulto_equivalente por hogar.

```

8.Por último, agreguen a respondieron una columna llamada pobre
que tome valor 1 si el ITF es menor al ingreso necesario que necesita
esa familia, y 0 en caso contrario. ¿Cuántos pobres identificaron para
cada año? ¿Qué porcentaje de la muestra representa?

```python
import pandas as pd

# 1) Cargar base
df = pd.read_csv("respondieron_p7.csv")

# 2) Asegurar numéricos
df["itf"] = pd.to_numeric(df.get("itf"), errors="coerce")
df["ingreso_necesario"] = pd.to_numeric(df["ingreso_necesario"], errors="coerce")

# 3) Crear variable pobre
df["pobre"] = (df["itf"] < df["ingreso_necesario"]).astype(int)

# 4) Resumen por año
resumen = (df.groupby("ano4")["pobre"]
             .agg(["sum","count"])
             .rename(columns={"sum":"n_pobres","count":"n_total"}))
resumen["pct_pobres"] = (resumen["n_pobres"] / resumen["n_total"] * 100).round(2)

# 5) Guardar base
df.to_csv("respondieron_p8.csv", index=False)
print("✔ Guardado respondieron_p8.csv")
print("\nResumen de pobreza por año:")
print(resumen)


```

comparando 2005 con 2025. Además, hagan 2 gráficos exploratorios a

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Cargar base con pobre
df = pd.read_csv("respondieron_p8.csv")

# 2) Tabla descriptiva por año
tabla_desc = df.groupby("ano4")["pobre"].describe().round(3)

print("=== Estadísticas descriptivas de pobreza por año ===")
print(tabla_desc)

# 3) Gráfico 1: Barras comparando % de pobreza entre 2005 y 2025
resumen = df.groupby("ano4")["pobre"].mean().reset_index()
resumen["pobre"] = resumen["pobre"] * 100

plt.figure(figsize=(6,4))
sns.barplot(data=resumen, x="ano4", y="pobre", palette="Set2")
plt.title("Porcentaje de hogares pobres (2005 vs 2025)")
plt.ylabel("% de pobres")
plt.xlabel("Año")
plt.tight_layout()
plt.show()

#grafico congruente con lo anterior.

# 4) Gráfico 2: Dispersión ITF vs Ingreso Necesario, coloreado por pobre
plt.figure(figsize=(7,5))
sns.scatterplot(data=df.sample(2000, random_state=1),  # muestra si el dataset es muy grande
                x="ingreso_necesario", y="itf", hue="pobre", alpha=0.6,
                palette={0:"green", 1:"red"})
plt.axline((0,0), slope=1, color="black", linestyle="--", label="Línea de corte")
plt.title("ITF vs Ingreso Necesario (pobre=1 en rojo)")
plt.xlabel("Ingreso Necesario")
plt.ylabel("ITF declarado")
plt.legend(title="¿Es pobre?")
plt.tight_layout()
plt.show()


```
